```python
import pandas as pd

df_population_raw = pd.read_csv('population_total.csv')
df_population_raw
```

```python
#dropping null values
df_population_raw.dropna(inplace=True)
```

```python
#making pivot table

df_pivot = df_population_raw.pivot(index= 'year', columns= 'country', values= 'population')
```

```python
#selecting some countries

df_pivot = df_pivot[['United States', 'China', 'India', 'Indonesia', 'Brazil']]
df_pivot
```

```python
#lineplot

df_pivot.plot(kind= 'line', xlabel= 'Year', ylabel= 'Population', title= 'Population (1955 - 2020)', figsize=(8,4))
```

```python
#barplot single year

df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
df_pivot_2020
```

```python
#rows to columns/ viceversa

df_pivot_2020 = df_pivot_2020.T
```

```python
#plotting barplot

df_pivot_2020.plot(kind='bar', xlabel= 'Country', ylabel= 'Population', title= 'Population 2020', color= 'red')
```

```python
#barplot multiple years

df_pivot_sample = df_pivot[df_pivot.index.isin([1990, 2000, 2010, 2020])]

df_pivot_sample.plot(kind= 'bar', xlabel= 'Country', ylabel= 'Population')
```

```python
#piechart

df_pivot_2020.rename(columns= {2020:'2020'}, inplace=True)

df_pivot_2020.plot(kind= 'pie', y= '2020', title= 'Population in 2020(%)')
```
